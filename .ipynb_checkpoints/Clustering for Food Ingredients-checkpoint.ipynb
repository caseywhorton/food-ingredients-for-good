# Clustering for food ingredients

**Purpose:** This notebook is intended to showcase clustering methods for food ingredients dataset.

Each observation in the dataset has a unique label (both a 'key', "NBD_No", and a text label for the ingredient, "Descrip"). My hypothesis is that similar foods have similar nutritional values, meaning we may be able to see distinct groupings of these foods based on their nutritional value.

I examine three (3) different clustering algorithms for this dataset:

1. [Kmeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)  
2. [Agglomerative Clustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html)
3. [DBSCAN](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

The K-means algorithm is usually my first algorithm to try when looking at clustering observations, and uses a Euclidean distance metric for assigning observations to clusters. The second and third algorithms have given better results in other projects I have been a part of. I wanted to test them for the case that there are a large number of clusters to assign observations to.

## General Approach

+ Take the first 5000 observations of the **cleaned_ingredients.csv** file/dataset
+ Take 10 samples of 1000 observations each and pass the sample to the clustering algorithm
+ For each sample, return:
    + [silhouette score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html) from each pass through the clustering algorithm
    + Other parameters such as cluster size
+ Plot the results

## Results and Next Steps

**Result:** There is some utility in clustering observations by their nutritional value, because some clusters have very similar foods assigned to them. However, this is not a flawless approach, since clusters will also contain very different foods, making it difficult to say what the cluster represents (in terms of food). Additionally, a numeric approach using the silhouette score for clustering algorithms shows that the clusters still overlap and are not distinct from each other.

```python
import pandas as pd
import numpy as np
from numpy import random


import matplotlib.pyplot as plt
import seaborn as sns

import os
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(os.getcwd() + '/data/cleaned_ingredients.csv')
```

```python
scaler = StandardScaler()
```

## K-Means Clustering

```python
df_list = []

for state in random.randint(0,100,100):
    df_sample = df.iloc[0:5000].sample(1000, random_state = state)
    X = np.array(df_sample.iloc[:,2:])
    X = scaler.fit_transform(X)

    scores = []
    cluster_sizes = range(5,30)
    
    for i in cluster_sizes:
        kmeans = KMeans(n_clusters=i, random_state=0).fit(X)
        scores.append(silhouette_score(X, kmeans.labels_))

    df_list.append(pd.DataFrame(data = {'cluster_sizes':cluster_sizes, 'scores':scores}))
    
dff = pd.concat(df_list)  
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(y = dff.scores, x = dff.cluster_sizes)
plt.xticks(rotation=45)
plt.title('kmeans - n=100')
plt.ylabel('silhouette scores')
plt.xlabel('cluster sizes')
plt.show()
```

**Interpretation:** There is a lot of variation in silhouette scores as we try to fit these food ingredients into more and more clusters. There is not much of a score gain by adding clusters, but if we try to maximize this score, maybe 12-15 could be selected and give decent results.

## Agglomerative Clustering

```python
df_list = []

for state in random.randint(0,100,100):
    df_sample = df.iloc[0:5000].sample(1000, random_state = state)
    X = np.array(df_sample.iloc[:,2:])
    X = scaler.fit_transform(X)

    scores = []
    cluster_sizes = range(5,30)
    
    for i in cluster_sizes:
        clustering = AgglomerativeClustering(n_clusters=i).fit(X)
        scores.append(silhouette_score(X, clustering.labels_))

    df_list.append(pd.DataFrame(data = {'cluster_sizes':cluster_sizes, 'scores':scores}))
    
dff = pd.concat(df_list)  

plt.figure(figsize=(10,5))
sns.boxplot(y = dff.scores, x = dff.cluster_sizes)
plt.xticks(rotation=45)
plt.title('agglomerative clustering - n=100')
plt.ylabel('silhouette scores')
plt.xlabel('cluster sizes')
plt.show()
```

**Interpretation:** Generally, the silhouette score increases as we add a number of clusters to group observations into. This is likely because the ingredients do not generally fit into a small number of clusters and there is a lot of overlap. For a cluster size of 17 (meaning we will have 17 clusters to fit observations into) the box plot shows the distribution of scores to be toward the 'peak'of the scores, while having the best lower quartile score.

## DBSCAN

```python
df_list = []

for state in random.randint(0,100,100):
    df_sample = df.iloc[0:5000].sample(3000, random_state = state)
    X = np.array(df_sample.iloc[:,2:])
    X = scaler.fit_transform(X)
    
    eps_list = [0.25, 0.5, 1, 2, 3]
    min_sample_list = [2, 3, 4, 5, 10]
    scores = []
    params = []
    sizes = []
    
    for i in eps_list:
        for j in min_sample_list:
            clustering = DBSCAN(eps=i, min_samples=j).fit(X)
            params.append((i,j))
            scores.append(silhouette_score(X, clustering.labels_))
            sizes.append(len(np.unique(clustering.labels_)))

    df_list.append(pd.DataFrame(data = {'params':params, 'scores':scores, 'sizes':sizes}))
    
dff = pd.concat(df_list)  
```

```python

sns.boxplot(y = dff.scores, x = dff.params)
plt.xticks(rotation=45)
plt.title('DBSCAN - n=100')
plt.ylabel('silhouette scores')
plt.xlabel('parameters')
plt.show()

sns.boxplot(y = dff.sizes, x = dff.params)
plt.xticks(rotation=45)
plt.title('DBSCAN - n=100')
plt.ylabel('number of clusters')
plt.xlabel('parameters')
plt.show()

```

## Examine Clusters

Utilizing the agglomerative clustering algorithm here, with parameter selection based on the chart above for agglomerative clustering, this section of the notebook looks at the clusters a little closer.

```python
df_sample = df.iloc[0:6000].sample(1000, random_state = 101)
X = np.array(df_sample.iloc[:,2:])
X = scaler.fit_transform(X)

clustering = AgglomerativeClustering(n_clusters=17).fit(X)
print('silhouette',(silhouette_score(X, clustering.labels_)))
```

```python
df_sample['label'] = clustering.labels_
for i in sorted(df_sample.label.unique()):
    print('*'*50)
    print('Label',i)

    print(df_sample[df_sample.label == i].Descrip.sample(20,replace=True).drop_duplicates())
```

Ingredients clustered into labels do not seem to group together ingredients into distinct clusters. 

For example, a single cluster with beans, vegetables, cereals and dairy would not be useful for categorizing foods together. Perhaps they have similar nutritional value though.

Some clusters do have consistent ingredients however, usually all cereal ingredients or all meat-based ingredients.

So, there is some utility in the clustering, but it is not perfect.

### Principal Components Analysis

```python
#PCA with two principal components
from sklearn.decomposition import PCA #Principal Component Analysis

pca_2d = PCA(n_components=2)

PCs_2d = pd.DataFrame(pca_2d.fit_transform(X))
PCs_2d.columns = ["PC1_2d", "PC2_2d"]
PCs_2d['label'] = clustering.labels_.astype(str)

plt.figure(figsize=(7,5))
sns.scatterplot(data = PCs_2d.sample(500), x = 'PC1_2d', y = 'PC2_2d', hue = 'label')
plt.title('PCA Agglomerative Clustering')
plt.show()
```

**Interpretaion:** There is some overlap in clusters based on our assignment of clusters. There are some outliers in their own clusters, which may explain the increase in the silhouette score by pushing some of the outliers into their own cluster instead of grouping them with the other observations. There is a large 'bunch' of foods that have similar component values, so we may need some other type of feature for these food ingredients to help set them apart.
